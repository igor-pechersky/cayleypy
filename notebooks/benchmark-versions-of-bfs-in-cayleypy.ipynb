```python
from cayleypy import prepare_graph, load_dataset, bfs_numpy, bfs_bitmask
import time
import pandas as pd
from IPython.display import display, HTML

GRAPHS = [("lrx",12), ("top_spin",12), ("pancake",11), ("full_reversals",10), ("all_transpositions",10)]
columns = ["Graph", "n", "BFS time", "bfs_numpy time", "bfs_bimtask time"]
table = []

for graph_name, max_n in GRAPHS:
    for n in range(4,max_n+1):
        correct_layer_sizes = load_dataset(graph_name + "_cayley_growth")[str(n)]
        graph = prepare_graph(graph_name, n=n)
        row = [graph_name, n]
        
        t0 = time.time()
        layer_sizes = graph.bfs().layer_sizes
        row.append("%.02fs" % (time.time() - t0))
        assert layer_sizes == correct_layer_sizes

        t0 = time.time()
        layer_sizes = bfs_numpy(graph)
        row.append("%.02fs" % (time.time() - t0))
        assert layer_sizes == correct_layer_sizes

        if n>8:
            t0 = time.time()
            layer_sizes = bfs_bitmask(graph)
            row.append("%.02fs" % (time.time() - t0))
            assert layer_sizes == correct_layer_sizes
        else:
            row.append("N/A")            

        print(row)
        table.append(row)

display(HTML(pd.DataFrame(table, columns=columns).to_html(index=False)))
```
