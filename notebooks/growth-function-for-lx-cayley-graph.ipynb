## Growth function for LX Cayley Graph

This notebook computes growth function for Cayley graph for symmetric group S_n when generated by two generators: left shift and swapping first two elements.

The sequence for diameters of this graph is https://oeis.org/A039745

```python
!pip install git+https://github.com/cayleypy/cayleypy
```

```python
from cayleypy import CayleyGraph, PermutationGroups, bfs_bitmask
import time

n=10

t0 = time.time()
graph = CayleyGraph(PermutationGroups.lx(n), verbose=2)
layers=bfs_bitmask(graph)
print("Layers:", layers)
print("diameter:", len(layers)-1)
print("Time: %.2fs" % (time.time()-t0))
```
