```python
from cayleypy import prepare_graph, load_dataset, CayleyGraph
import time
import pandas as pd
from IPython.display import display, HTML

GRAPHS = [("full_reversals",12), ("all_transpositions",12), ("coxeter",12), ("lrx",12), ("top_spin",12), ("pancake",12), ]
columns = ["Graph", "n", "BFS time (GPU)"]
table = []

for graph_name, max_n in GRAPHS:
    for n in range(4,max_n+1):
        correct_layer_sizes = load_dataset(graph_name + "_cayley_growth")[str(n)]
        graph_def = prepare_graph(graph_name, n=n)
        graph = CayleyGraph(graph_def, device="cuda", memory_limit_gb=16, batch_size=2**20)
        graph.free_memory()
        row = [graph_name, n]
        
        t0 = time.time()
        layer_sizes = graph.bfs().layer_sizes
        row.append("%.02fs" % (time.time() - t0))
        assert layer_sizes == correct_layer_sizes      

        print(row)
        table.append(row)

display(HTML(pd.DataFrame(table, columns=columns).to_html(index=False)))
```
